### Visualización de datos

Hay una serie de funciones para importar datos, el primer paso en el  análisis de datos. Por lo general, es una buena práctica mirar los datos antes de ponerlos  en un modelo, por lo que es lo que vamos a hacer en el siguiente paso. La razón es simple: diferentes modelos tienen diferentes puntos fuertes y ningún modelo es universalmente la mejor opción para todos los casos, ya que tienen un conjunto diferente de suposiciones. Aplicar arbitrariamente un modelo sin comprobar los datos con respecto a sus supuestos suele dar lugar a conclusiones engañosas.

Una manera inicial de elegir un modelo y realizar tales verificaciones es simplemente examinar visualmente los datos mirando sus límites y patrones. En otras palabras, necesitamos visualizar los datos primero. 

#### Creación de diagramas de dispersión

En `R`, la función básica para visualizar datos es `plot()`. Si simplemente proporcionamos un vector numérico o entero a `plot()`, producirá un gráfico de dispersión de valor por índice. Por ejemplo, el siguiente código crea un diagrama de dispersión de `10` puntos en orden creciente:

````r
plot(1:10)
````

Podemos crear un diagrama de dispersión más realista generando dos vectores numéricos aleatorios linealmente correlacionados:

````r
x <- rnorm(100)
y <- 2 * x + rnorm(100)
plot(x, y)
````

####  Personalización de  los  elementos de un  gráfico

En un gráfico, hay numerosos elementos de un  gráfico que se pueden personalizar. Los elementos más comunes a especificar son el título (`title()`), la etiqueta del eje `x` (`xlab`), la etiqueta del eje `y` (`ylab`), el rango del eje `x` (`xlim`) y el rango del eje `y` (`ylim`):

````r
plot(x, y,
     main = "Números aleatorios linealmente correlacionados",
     xlab = "x", ylab = "2x + ruido",
     xlim = c(-4, 4), ylim = c(-4, 4))
````

El título del gráfico se puede especificar mediante el argumento principal o  por una llamada separada de una función `title()`. Por lo tanto, el código anterior es equivalente al código siguiente:

```
plot(x, y,
    xlim = c(-4, 4), ylim = c(-4, 4),
    xlab = "x", ylab = "2x + ruido")
    title("Numeros aleatorios linealmente correlacionados")
```

#### Personalización de estilos de puntos

El estilo de puntos predeterminado de un gráfico de dispersión es un círculo. Al especificar el argumento `pch` , podemos cambiar el estilo de punto. Hay `26` estilos de puntos disponibles:

````r
plot(0:25, 0:25, pch = 0:25,
     xlim = c(-1, 26), ylim = c(-1, 26), 
     main = "Estilo de puntos (pch)")
text(0:25+1, 0:25, 0:25)
````

El código anterior produce un diagrama de dispersión de todos los estilos de puntos disponibles, con los números  correspondientes de `pch`  impresos al lado. Primero, `plot()` crea un diagrama de dispersión simple, y luego `text()` imprime los números `pch` en el lado derecho de cada punto.

Al igual que muchas otras funciones incorporadas, `plot()` se vectoriza con respecto a `pch` y varios otros argumentos. Esto permite personalizar el estilo de cada punto en el diagrama de dispersión. Por ejemplo, el caso más simple es que usamos sólo un estilo de punto no predeterminado para todos los puntos, estableciendo `pch = 16`:

````r
x <- rnorm(100)
y <- 2 * x + rnorm(100)
plot(x, y, pch = 16,
     main = "Diagrama de dispersión con estilo de punto personalizado")
````

A veces, necesitamos distinguir dos grupos de puntos por una condición lógica. Sabiendo que `pch` está vectorizado, podemos usar `ifelse()`, para especificar el estilo de punto de cada observación, examinando si un punto satisface la condición. En el siguiente ejemplo, queremos aplicar `pch = 16` a los puntos que satisfacen `x * y > 1` de lo contrario, `pch = 1`:

````r
plot(x, y,
pch = ifelse(x * y > 1, 16, 1),
     main = "Gráfico de dispersión con estilos de punto condicional")
````

También podemos dibujar los puntos en dos conjuntos de datos separados que comparten el mismo eje `x` usando `plot()` y `points()`. En el ejemplo anterior, se generó un vector aleatorio `x` distribuido normalmente y un vector aleatorio linealmente correlacionado `y`.

Ahora, generamos otro vector aleatorio `z` que tiene una relación no lineal con `x` y dibuja  `y` e `z` vs  `x`  pero con diferentes estilos de puntos:

````r
z <- sqrt(1 + x ^ 2) + rnorm(100)
plot(x, y, pch = 1,
     xlim = range(x), ylim = range(y, z),
     xlab = "x", ylab = "valor")
points(x, z, pch = 17)
title("Gráfico de dispersión con dos series")
````

Después de generar z, creamos un gráfico de `x` e `y` primero y luego agregamos otro grupo de puntos `z` con un `pch` diferente. Tenga en cuenta que si no especificamos `ylim = range(y, z)`, el constructor del gráfico  sólo considerará el rango de `y`  y el eje `y` puede tener un rango más estrecho que el rango de `z`.  Desafortunadamente, `points()` no alarga automáticamente los ejes creados por `plot()`, por lo tanto cualquier punto más allá del rango de los ejes desaparecerá. 

El código anterior establece un rango apropiado del eje `y`, para que todos los puntos en `y` e `z` puedan ser mostrados en el área del gráfico.

#### Personalización de  colores de puntos

Si los gráficos no se limitan a la visualización  a escala de grises,  podemos usar colores diferentes de puntos al establecer la columna de `plot()`:

````r
plot(x, y, pch = 16, col = "blue",
     main = "Gráfico de dispersión con puntos azules")
````

Como `pch`, `col` es también un argumento vectorial. Con el mismo método, podemos aplicar diferentes colores para separar puntos en dos categorías diferentes dependiendo de si satisfacen una determinada condición:

````r
plot(x, y, pch = 16, 
     col = ifelse(y >= mean(y), "red", "green"),
     main = "Gráfico de dispersión con puntos de colores condicional")
````

Tenga en cuenta que si el diagrama de dispersión se imprime en escala de grises, los colores sólo se pueden ver como diferentes intensidades de gris. Además, podemos usar `plot()` y `points()` de nuevo, pero con diferentes `col` para  distinguir diferentes grupos de puntos:

````r
plot(x, y, col = "blue", pch = 0,
     xlim = range(x), ylim = range(y, z),
     xlab = "x", ylab = "valor")
points(x, z, col = "red", pch = 1)
title("Gráfico de dispersión con dos series ")
````

#### Creación de gráficos de línea

Para los datos de series de tiempo, las gráficas de línea son más útiles para demostrar la tendencia y la variación a través del tiempo. Para crear gráficos de líneas, solo necesitamos establecer `type = "l"` al llamar a `plot()`:

````r
t <- 1:50
y <- 3 * sin(t * pi / 60) + rnorm(t)
plot(t, y, type = "l",
     main = "Gráfico de línea simple")
````

#### Personalización del tipo y ancho de línea

Al igual que `pch` para el diagrama de dispersión, `lty` se utiliza para especificar el tipo de línea de un gráfico de líneas. A continuación se muestra una vista previa de seis tipos de líneas compatibles con R:

````r
lty_valores <- 1:6
plot(lty_valores, type = "n", axes = FALSE, ann = FALSE)
abline(h =lty_valores, lty = lty_valores, lwd = 2)
mtext(lty_valores, side = 2, at = lty_valores)
title("Tipo de líneas (lty)")
````

El código anterior crea un cuadro vacío con `type = "n"` con rangos de ejes apropiados, otra etiqueta `elements.abline()` se utiliza para dibujar las líneas horizontales con diferentes tipos de línea pero de igual ancho de línea (`lwd = 2` ). La función `mtext()` se utiliza para dibujar el texto en el margen. Ten en cuenta que `abline()` y `mtext()` se vectorizan con respecto a sus argumentos de modo que no necesitamos un bucle `for` para dibujar cada línea y  el margen de texto a su vez.

El siguiente ejemplo demuestra cómo `abline()` puede ser útil para dibujar líneas auxiliares en un gráfico. Primero, creamos un gráfico de línea de `y` con tiempo `t`,  como se definió en el ejemplo de gráfico de líneas. Supongamos  también queremos que el gráfico muestre el valor medio y el intervalo de `y` junto con el valor de `t` en que aparecen los valores máximo y mínimo. Con `abline()`, podemos dibujar fácilmente estas líneas auxiliares con diferentes tipos de líneas y colores para evitar la ambigüedad:

````r
plot(t, y, type = "l", lwd = 2)
abline(h = mean(y), lty = 2, col = "blue")
abline(h = range(y), lty = 3, col = "red")
abline(v = t[c(which.min(y), which.max(y))], lty = 3, col = "darkgray")
title("Gráfico de línea con líneas auxiliares")
````

### Gráfico  de líneas en varios períodos

Otro tipo de gráfico de línea en el que se mezclan diferentes tipos de líneas es un gráfico de líneas de varios periodos. Una forma típica es que el primer período es de datos históricos y el segundo período es de predicciones. Supongamos que el primer período de `y` incluye las primeras `40` observaciones y los puntos restantes son predicciones basadas en los datos históricos. Queremos utilizar líneas sólidas para representar los datos históricos y líneas discontinuas para las predicciones.

Se debe tener  en cuenta que `lines()` es a un gráfico de línea como `points()` es a un gráfico de dispersión:

````r
p <- 40
plot(t[t <= p], y[t <= p], type = "l",
xlim = range(t), xlab = "t")
lines(t[t >= p], y[t >= p], lty = 2)
title("Gráfico de línea simple con dos periodos")
````

#### Dibujando  líneas con puntos

A veces, es útil dibujar líneas y puntos en el mismo gráfico para enfatizar que las observaciones son discretas o simplemente hacer que el gráfico sea más claro. El método es simple, sólo dibujamos un gráfico de líneas y agregamos `points()` de los mismos datos al gráfico :

````r
plot(y, type = "l")
points(y, pch = 16)
title("Líneas con puntos")
````

Una forma equivalente de hacerlo es dibujar un diagrama de dispersión primero usando la función `plot()` y luego agregar líneas usando la función `lines()` de los mismos datos al gráfico nuevamente. Por lo tanto, el código siguiente debería producir exactamente los mismos gráficos que el ejemplo anterior:

````r
plot(y, pch = 16)
lines(y)
title("Líneas con puntos")
````

#### Gráfico de varias series con una leyenda

La versión completa de un gráfico de varias series debe incluir varias series representadas por líneas y puntos, y una leyenda para ilustrar las series en el gráfico. El siguiente código genera aleatoriamente dos series, `y` e  `z`, con el tiempo `x` y crea un gráfico con estos conjuntos:

````r
x <- 1:30
y <- 2 * x + 6 * rnorm(30)
z <- 3 * sqrt(x) + 8 * rnorm(30)
plot(x, y, type = "l",
     ylim = range(y, z), col = "black")
points(y, pch = 15)
lines(z, lty = 2, col = "blue")
points(z, pch = 16, col = "blue")
title ("Gráfico de dos  series")
legend("topleft",
legend = c("y", "z"),
col = c("black", "blue"),
lty = c(1, 2), pch = c(15, 16),
cex = 0.8, x.intersp = 0.5, y.intersp = 0.8)
````

El código anterior usa `plot()`, para crear el gráfico de puntos de línea de `y`, luego se agrega `lines()` y los `points()` de `z`. Al final, agregamos `legend()` en la parte superior izquierda para demostrar los estilos de línea y punto de `y`  e ` z`, respectivamente. Se debe tener en cuenta que `cex` se utiliza para escalar los tamaños de fuente de la leyenda y `x.intersp` e `y.intersp` se utilizan para ajustes menores a la leyenda.


#### Creación de gráficos de barras

En las secciones anteriores, se aprendió a crear diagramas de dispersión y gráficos de líneas. Hay varios otros tipos de gráficos que son útiles y dignos de mención. El gráfico de barras  están entre las más usados. La altura de las barras en un gráfico de barras pueden hacer un contraste cuantitativamente entre diferentes categorías.

El gráfico de barras más simple que podemos crear es el siguiente. Aquí, usamos `barplot()` en lugar de `plot()`:

````r
barplot(1:10, names.arg = LETTERS[1:10])
````

Si el vector numérico tiene nombres, los nombres serán automáticamente   los nombres en el eje `x`. Por lo tanto, el código siguiente producirá exactamente el mismo gráfico de barras que lo anterior:

````r
ints <- 1:10
names(ints) <- LETTERS[1:10]
barplot(ints)
````

La elaboración de un gráfico de barras parece tan fácil. Ahora que tenemos el dataset de vuelos en `nycflights13`, podemos crear un diagrama de barras de las ocho principales aerolíneas con más vuelos en el registro:

````r
#install.packages("nycflights13")
library("nycflights13")
data("flights", package = "nycflights13")
aerolineas <- table(flights$carrier)
aerolineas
````

En el código precedente, `table()` se utiliza para contar el número de vuelos en el registro para cada portador:

````r
ordenada_aerolineas <- sort(aerolineas, decreasing = TRUE)
ordenada_aerolineas
````

Como se muestra en el código anterior, las portadores se ordenan en orden descendente. Podemos sacar los primeros 8 elementos de la tabla y hacer un gráfico de barras:

````r
barplot(head(ordenada_aerolineas, 8),
        ylim = c(0, max(ordenada_aerolineas) * 1.1),
        xlab = "Aerolineas", ylab = "Vuelos",
        main ="Las 8 principales aerolíneas con más vuelos registrados")
````

Otro gráfico útil son los gráficos circulares. La función para crear un gráfico circular, es `pie()`, funciona de manera similar a `barplot()`. #### Histogramas  y gráficos de  densidad

Los gráficas de dispersión y los  gráficos  de línea son ilustraciones directas de las observaciones de un conjunto de datos. Los gráficos de barras y los gráficos circulares se usan generalmente para mostrar un resumen aproximado de los puntos de datos en  diferentes categorías.

Los diagramas de dispersión y los gráficos  de línea transmiten demasiada información y son difíciles de extraer información, mientras que los gráficos de barras y los gráficos circulares dejan demasiada información, por lo que  también puede ser difícil hacer un juicio concluyente con confianza a partir de estos gráficos .

Un histograma muestra la distribución de un vector numérico y resume la información en los datos sin dejar demasiada información  y por lo tanto puede ser más fácil de manejar. El siguiente ejemplo muestra cómo utilizar `hist()` para producir un histograma de un vector numérico aleatorio normalmente distribuido y la función densidad de la distribución normal:

````r
random_normal <- rnorm(10000)
hist(random_normal, main = "Histograma de un vector normal", ylab = "Frecuencia")
````

De forma predeterminada, el eje `y` de un histograma es la frecuencia de los valores en los datos. Podemos comprobar que el histograma se acerca a una distribución normal estándar a partir de la cual se generó `random_normal`. Para superponer la curva de una función densidad de probabilidad de la distribución normal estándar, `dnorm()`, debemos asegurarnos de que el eje `y` del histograma es una probabilidad y la curva debe añadirse al histograma:

````r
hist(random_normal, main = "Histograma de un vector normal", ylab = "Densidad",  probability = TRUE, col = "lightgray")
curve(dnorm, add = TRUE, lwd = 2, col ="blue")
````

Ahora, hagamos un histograma de la velocidad de un avión en vuelo. Básicamente, la velocidad media de un avión en un viaje es la distancia del viaje (`distance`) dividida por el tiempo en el aire (`air_time`):

````r
velocidad_vuelo <- flights$distance / flights$air_time
hist(velocidad_vuelo, main = "Histograma de las velocidades de vuelo", xlab = "Velocidad de vuelo", ylab = "Frecuencia")
````

El histograma parece un poco diferente de una distribución normal. En este caso, utilizamos la `density()` para estimar una distribución empírica de la velocidad, dibujamos  una curva de distribución de probabilidad bastante suave fuera de ella y añadimos  una línea vertical para indicar el promedio global de todas las observaciones:

````r
plot(density(velocidad_vuelo, from = 2, na.rm = TRUE),
     main ="Distribución empírica de la velocidad de vuelo ", ylab ="Densidad")
abline(v = mean(velocidad_vuelo, na.rm = TRUE),
col = "blue", lty = 2)
````

Podemos combinar los dos gráficos juntos para obtener una mejor vista de los datos:

````r
hist(velocidad_vuelo, probability = TRUE, ylim = c(0, 0.5),
     main ="Histograma y distribución  empiríca de velocidad de vuelo",
     border ="gray", col = "lightgray",  ylab ="Densidad", xlab = "Velocidad de vuelo")
lines(density(velocidad_vuelo, from = 2, na.rm = TRUE),
      col ="darkgray", lwd = 2)
abline(v = mean(velocidad_vuelo, na.rm = TRUE),
       col ="blue", lty =2)
````

#### Creación de un diagrama de  cajas

Los histogramas y las gráficas de densidad son dos maneras de demostrar la distribución de los datos. Por lo general, sólo necesitamos varios quantiles críticos para obtener una impresión  de toda la distribución. Los diagrama  de cajas es una manera simple de hacer esto. Para un vector numérico generado aleatoriamente, podemos llamar a `boxplot()` para dibujar un diagrama de cajas:

````r
x <- rnorm(1000)
boxplot(x)
````

El código siguiente dibuja un diagrama de cajas de la velocidad de vuelo para cada aerolinea. Habrá 16 cajas en un gráfico, lo que facilita la comparación aproximada de la distribución de las diferentes aerolineas.

Para proceder, utilizamos la fórmula `distance/air_time ~ carrier` para indicar que el eje `y` indica la velocidad de vuelo calculada a partir de  `distance/air_time` y el eje `x`  indica la aerolínea.

````r
boxplot(distance / air_time ~ carrier, data =flights,main = "Diagrama  de cajas de la velocidad de vuelo por aerolínea")
````

Se debe tener  en cuenta que usamos la interfaz de la fórmula para crear gráficos en `boxplot()`. En este caso, `distance/air_time~carrier`,  básicamente significa que el eje `y` debe representar los valores de `distance/air_time`, es decir, las velocidad de vuelo  y el eje `x` debe representar diferentes aerolíneas. `data = flights` le dice a `boxplot()` dónde encontrar los símbolos en la fórmula que especificamos. Como resultado, el diagrama de cajas de la velocidad de vuelo se crea y se agrupa por las aerolíneas.
